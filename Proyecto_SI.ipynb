```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
dataset = pd.read_csv('/content/cardio.csv', sep=";")
dataset
```

```python
dataset = dataset.values
dataset
```

```python
X = dataset[:, 0:12]
Y = dataset[:, 12]
model = Sequential()
model.add(Dense(12,input_dim=12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=['accuracy'])
model.fit(X,Y,epochs=150,batch_size=10)
```

```python
from numpy.lib.npyio import savez_compressed
score = model.evaluate(X,Y)
print(model.metrics_names[1],score[1]*100)
```

```python
prediccion = model.predict(X)
prediccion = [1 if y>0.5 else 0 for y in prediccion]
print(prediccion)
```

```python
respuesta = model.predict([[3,176232,169,82.0,150,100,1,1,0,0,1,1]])
print(1 if respuesta>0.5 else 0)
```

```python
model_json = model.to_json()
with open("model.json","w") as json_file:
  json_file.write(model_json)
model.save_weights("model.h5")
print("modelo guardado")
```
